## 1º projeto: modelode classificação

Será classificado atravez do modelo e definido se o animal é porco ou chachorro


Para que a maquina defina se  é um porco ou cachorro temos que criar caracteristicas para classificação, detalhes que devem ter em um  porco e um cachorro para que a maquina possa apreende de melhor forma


```python
import pandas as pd
import numpy as np

np.random.seed(5)

porco_1 = [0, 1, 0]
porco_2 = [0, 1, 1]
porco_3 = [1, 1, 0]

cachorro_1 = [0, 1, 1]
cachorro_2 = [1, 0, 1]
cachorro_3 = [1, 1, 1]

index = ["porco1", "porco2", "porco3", 'cachorro1','cachorro2','cachorro3']
columns = ['pelo_curto', 'perna_curta', 'late']

train_x = [ porco_1, porco_2, porco_3, cachorro_1, cachorro_2, cachorro_3]
train_y = [ 1, 1, 1, 0, 0, 0]
```

## Definições

para o modelo criado abaixo os valores 0 é para cachorro e valores 1 serão associadosa porcos

```python
df_modelo_dados = pd.DataFrame(columns=columns, data=train_x, index=index)
df_modelo_dados['Classificação'] = train_y
df_modelo_dados
```

# Modelo SUPERVISIONADO por Classificações

Nesse modleo para o apreendizagem de maquina serão informado algusn dados pré estimados
Esses  dados devem contar FEATURES ou base para decisões e a CLASSES que são o retorno das decisões

Uma vez classificado edeterminado a FEATURES de um modelo então podemos definar retorno classificado e usar isso para alimentar o algoritmo que vai decidir o retorno dos dados.

Além de que precisamos verrifica a ACCURACIA do retorno desses dados para validar a precisão do nosso modelo de machine learn


```python
## importa a biblioteca que será trabalhada nesse projeto

from sklearn.svm import LinearSVC

model =  LinearSVC()

model.fit(train_x,train_y)

```

```python
## O MODELO PREDICT VERIFICA COM BASE EM NOVAS FEATURES QUAL SERÁ
## A SAÍDA PARA NOVAS CLASSES

animal_misteriosos = [ 1, 1, 1]

model.predict([animal_misteriosos])
```

```python
misterio_1 = [1, 1, 1]
misterio_2 = [1, 1, 0]
misterio_3 = [0, 1, 1]

test_x = [misterio_1, misterio_2, misterio_3]
test_y = [ 0, 1, 1]

teste_predict = model.predict(test_x)

acertos = (teste_predict == test_y).sum()
total = len(teste_predict)
taxa_acerto = ((acertos/total)*100).round(2)

print(f" a taxa de acerto para o modelo foi {taxa_acerto} (%)")
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto = accuracy_score(test_y, teste_predict)
print(f" A taxa de acerto foi { (taxa_acerto*100).round(2)} (%)")
```

```python
import pandas as pd

uri = 'https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv'
dados = pd.read_csv(uri)
```

### TRANSFORMANDO DADOS PARA TRABALHAR
* RENOMEAR COLUMNS
* AJUSTAR DADOS PARA BINÁRIOS

```python
dados.head()
```

```python
new_columns_name = {
    'mileage_per_year': 'milhas_por_ano',
    'model_year': 'ano_do_modelo',
    'price': 'preco',
    'sold': 'vendido'
}

dados = dados.rename(columns=new_columns_name)
dados.head()
```

```python
new_data_vendido = {
    'yes': 1,
    'no': 0
}

dados['vendido'] = dados['vendido'].map(new_data_vendido)
dados.head()
```

```python
from datetime import datetime

ano_atual = datetime.today().year
ano_atual
```

```python
dados['ano_do_veículo'] = ano_atual - dados['ano_do_modelo']
dados.head()
```

```python
dados['km_por_ano'] = dados['milhas_por_ano'] * 1.60934
dados.head()
```

```python
dados = dados.drop(['milhas_por_ano', 'ano_do_modelo', 'Unnamed: 0'], axis=1)

dados.head()
```

```python
x = dados[['ano_do_veículo', 'km_por_ano', 'preco']]
y = dados['vendido']
```

```python
x.head()
```

```python
y.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


SEED = 20
np.random.seed(SEED)

raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y,
                                                    stratify=y,
                                                    random_state=SEED,
                                                    train_size=0.75)

scaler = StandardScaler()
scaler.fit(raw_train_x)
train_x = scaler.transform(raw_train_x)
test_x = scaler.transform(raw_test_x)

model = LinearSVC()
model.fit(train_x, train_y)
prediction = model.predict(test_x)

accuracy = (accuracy_score(test_y, prediction) * 100).round(2)

print(f'Para o modelo de apredizagem de carros a taxa de accuracy foi de {accuracy} (%)')

```

## DUMMY CLASSIFIER

Classificadores bodos


```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()

dummy.fit(train_x, train_y)
prediction = dummy.predict(test_x)

dummy_accuracy = dummy.score(test_x, test_y)

accuracy = (accuracy_score(test_y, prediction)* 100).round(2)

print(f'O teste com Dummy deu uma accuracy de {accuracy} (%)')
print(f'O teste com Dummy Score deu uma accuracy de {dummy_accuracy} (%)')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


SEED = 20
np.random.seed(SEED)

raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y,
                                                    stratify=y,
                                                    random_state=SEED,
                                                    train_size=0.75)

model = DecisionTreeClassifier(max_depth=3)
model.fit(raw_train_x, train_y)
prediction = model.predict(raw_test_x)

accuracy = (accuracy_score(test_y, prediction) * 100).round(2)

print(f'Para o modelo de apredizagem de carros a taxa de accuracy foi de {accuracy} (%)')
```
